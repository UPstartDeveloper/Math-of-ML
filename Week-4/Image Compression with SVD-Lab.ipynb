# Perform SVD Decompostion

```python
import os

from matplotlib.image import imread
import matplotlib.pyplot as plt
import numpy as np
from  numpy.linalg import matrix_rank as rank
```

## Helper Functions

```python
def image_show(X, title):
    img = plt.imshow(X)
    img.set_cmap('gray')  # default behavior is to add color, even if grayscale :)
    plt.axis("off")
    plt.title(title)
    plt.show() 
 
def rbg_gray(filename):
    A = imread(filename)
    grayed = np.mean(A, -1)  # just a way of smushing the RGB channels into 1
    return grayed
```

```python
filename = "sanFran.png"
grayscale_img = rbg_gray(filename)

# Show the image
image_show(grayscale_img, "Original Image")
```

```python
#Show different reduced version of the image
image_show(
    grayscale_img, 
    "Rank r = 327"
)
```

```python
U, S, VT = np.linalg.svd(grayscale_img)
S = np.diag(S)
```

```python
for rank in [rank(grayscale_img), 100, 50, 25, 10, 5, 1]:
    r = rank
    X_approx = U[:, :r] @ S[:r, :r] @ VT[:r, :]
    image_show(X_approx, f"Rank r = {r}")
```
